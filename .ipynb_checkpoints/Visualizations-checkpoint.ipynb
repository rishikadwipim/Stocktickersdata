```python
from cassandra.cluster import Cluster
from cassandra.auth import PlainTextAuthProvider
import json

# Path to your secure connect bundle and token JSON file
secure_connect_bundle_path = "secure-connect-stockdata.zip"
token_file_path = "StockData-token.json"

# Load the token file to extract credentials
with open(token_file_path) as f:
    secrets = json.load(f)

# Extract the client ID and client secret from the token
CLIENT_ID = secrets["clientId"]
CLIENT_SECRET = secrets["secret"]

# Define the authentication provider using the extracted credentials
auth_provider = PlainTextAuthProvider(CLIENT_ID, CLIENT_SECRET)

# Set up the cluster connection using the secure connect bundle and authentication provider
cloud_config = {
    'secure_connect_bundle': secure_connect_bundle_path
}

# Create a Cassandra cluster connection
cluster = Cluster(cloud=cloud_config, auth_provider=auth_provider)
session = cluster.connect()

session.set_keyspace('tickers')

# Verify the connection by executing a simple query (optional)
row = session.execute("select release_version from system.local").one()
if row:
    print(f"Cassandra Version: {row[0]}")
else:
    print("An error occurred.")


```

```python
import pandas as pd
# Replace with your Cassandra keyspace

# Query to fetch data from the gold table
query = "SELECT base_currency_name, currency_name, market, total_pairs FROM gold1_currency_pairs_summary LIMIT 10"
rows = session.execute(query)

# Convert the result to a Pandas DataFrame
df1 = pd.DataFrame(rows)

# Check the DataFrame
df1.head()
```

```python
# Create a pie chart for market distribution
market_distribution = df1['market'].value_counts()

plt.figure(figsize=(7, 7))
market_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['#f48fb1', '#64b5f6'])
plt.title('Market Distribution of Currency Pairs')
plt.ylabel('')  # Remove the ylabel for better appearance
plt.show()
```

```python
# Pivot the data to get counts of total pairs for each currency by market
market_currency_pairs = df1.pivot_table(values='total_pairs', index='currency_name', columns='market', aggfunc='sum', fill_value=0)

# Create a stacked bar chart
market_currency_pairs.plot(kind='bar', stacked=True, figsize=(12, 8), color=['#f48fb1', '#64b5f6'])
plt.title('Currency Pairs by Market and Currency')
plt.xlabel('Currency Name')
plt.ylabel('Total Pairs')
plt.xticks(rotation=90)
plt.legend(title='Market', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from cassandra.cluster import Cluster
import pandas as pd
import matplotlib.pyplot as plt


# Step 3: Query the table
query = "SELECT currency_name, total_pairs FROM gold1_currency_pairs_summary;"
rows = session.execute(query)

# Step 4: Load data into a Pandas DataFrame
data = pd.DataFrame(rows)

# Step 5: Aggregate values using Pandas
currency_counts = data.groupby('currency_name')['total_pairs'].sum().reset_index()

# Step 6: Sort the data and exclude the first position
top_10_excluding_first = currency_counts.sort_values(by='total_pairs', ascending=False).iloc[1:11]

# Step 7: Plot the data
plt.figure(figsize=(12, 8))
plt.bar(top_10_excluding_first['currency_name'], top_10_excluding_first['total_pairs'], color='skyblue')
plt.xlabel('Currency Name', fontsize=12)
plt.ylabel('Total Pairs', fontsize=12)
plt.title('Top 10 Currencies by Total Pairs (Excluding 1st Position)', fontsize=16)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

```
